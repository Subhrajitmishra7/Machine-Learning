# Steps

```python
#1. Prepare the data
x=[1,2,3,4,5]
y=[1,4,9,16,25]

#2.Setup plot
fig,ax = plt.subplots(figsize=(5,5))#(width,height)

#3.Plot data
ax.plot(x,y)

#4.Customizing the plot
ax.set(title="Numbers and their square",
      xlabel = "Numbers",
      ylabel = "Squares")
```

```python
#Making figures using Numpy arrays

#Creating some data
x=np.linspace(10,100,10)
#linespace has a starting point,a end point and the third parameter depicts that many numbers between the starting point and the ending point
#Plot the data and the line plot
fig ,ax = plt.subplots()
ax.plot(x,x**2);
```

```python
#Creating a scatter plot on the same data
fig,ax = plt.subplots()
ax.scatter(x,x**2);
```

```python
#Creating a line plot on the same data
fig,ax = plt.subplots()
ax.plot(x,np.sin(x))
```

```python
#Creating a bar plot from a dictonary
dict = {"Nike":10000,"Adidas":8000,"Puma":9000,"Fila":7000,"Asics":5000}
fig,ax=plt.subplots()
ax.bar(dict.keys(),dict.values())
plt.title("Shoes-price bar graph")
plt.xlabel("Brand name")
plt.ylabel("Price ")
```

```python
#Creating a horizontal bar graph
dict = {"Nike":10000,"Adidas":8000,"Puma":9000,"Fila":7000,"Asics":5000}
fig, ax=plt.subplots()
#There is a cache doing it,for horizontal demonstration the keys and values must be converted into lists
ax.barh(list(dict.keys()),list(dict.values())
```

```python
names=["Nike","Adidas","Puma","Fila","Asics"]
prices=[10000,8000,9000,7000,5000]
fig,ax=plt.subplots()
ax.barh(names,prices)
ax.set(title="Shoe-price Graph",
      xlabel="Prices",
      ylabel="Brand Name")
```

# Creating  subplots inside a subplot

```python
#Option 1 of creating subplots
fig, ((ax1,ax2),(ax3,ax4)) = plt.subplots(nrows=2,ncols=2,figsize=(8,8))
x=[1,2,3,4]
ax1.scatter(x,x)
ax2.plot(x,x)
ax3.barh(names,prices)
ax4.bar(dict.keys(),dict.values())
```

```python
#Option 2 of creating subplots
fig,ax = plt.subplots(nrows=2,
                     ncols=2,
                     figsize=(8,8))
ax[0,0].scatter(x,x)
ax[0,1].plot(x,x)
ax[1,0].barh(names,prices)
ax[1,1].bar(dict.keys(),dict.values())
```

# # Which one should yu use?(pyplot vs matplotlib OO method)

#-> when plotting something quickly prefer to use the pyplot method
#-> when plotting somrthing advanced use the OO method

```python
heart_diesease.head()
```

```python
over_50 = heart_diesease[heart_diesease["age"]>50]
over_50.head(50)
```

```python
#Pyplot method
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target",
            figsize=(8,8))
```

```python
over_50.plot(kind="bar",
            x="age",
            y="chol",
            figsize=(8,8))
```

```python
over_50.plot(kind="hist",
            x="age",
            y="chol",
            figsize=(8,8))
```

```python
##Creating using the OO method mixed with Pyplot method
fix,ax=plt.subplots(figsize=(8,8))
over_50.plot(kind="scatter",
            x="age",
            y="chol",
            c="target", 
            ax=ax);
```

```python
#OO method implmentation from the scratch
fig,ax=plt.subplots(figsize=(8,8))
#Prepare the plot
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"]);

#Customize the plot
ax.set(title="Heart Diesease and Cholestrol Level",
      xlabel="Age",
      ylabel="Cholestrol");

#Adding the legends
ax.legend(*scatter.legend_elements(),title="target");

#Adding a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
over_50
```

```python
fig,ax=plt.subplots(figsize=(8,8))

#Creating the data
scatter=ax.scatter(x=over_50["age"],
                  y=over_50["thalach"],
                  c=over_50["target"])

#Creating a horizontal line
ax.axhline(over_50["thalach"].mean(),
           linestyle='--');

#Creating legend
ax.legend(*scatter.legend_elements(),title="target");

#Customizing the plot
ax.set(title="Age vs Heart Rate Analysis",
      xlabel="Age",
      ylabel="Heart Rate");
```

```python
#Subplot of age-thalach and age-chol together
fig,(ax0,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)
#sharex attribute help in all the subplots to share the same axes

#Providing data to the subplots
#Providing data to the ax0
scatter1=ax0.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"])
#Customizing the ax0
ax0.set(title="Heart Diesease vs Thalach",
       xlabel="Heart Diesease",
       ylabel="Thalach")
#Setting the legend
ax0.legend(*scatter1.legend_elements(),title="target")

#Customizing mean Thalach line
ax0.axhline(over_50["thalach"].mean(),
           linestyle="--")


#Providing data to the subplots
#Providing data to the ax1
scatter2=ax1.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"])

#Customizing the ax1
ax1.set(title="Heart Diesease vs cholestrol",
       xlabel="Heart Diesease",
       ylabel="cholestrol")

#Setting the legend
ax1.legend(*scatter2.legend_elements(),title="target")

#Customizing mean cholestrol line
ax1.axhline(over_50["chol"].mean(),
           linestyle="--");

#Add title to the plot
fig.suptitle("Heart Disease Analysis",fontsize=18,fontweight=20);
```

# Customizing the plots

```python
#Setting style to our plots
plt.style.use("seaborn-whitegrid")
#OO method implmentation from the scratch
fig,ax=plt.subplots(figsize=(8,5))
#Prepare the plot
scatter = ax.scatter(x=over_50["age"],
                    y=over_50["chol"],
                    c=over_50["target"],
                    cmap="plasma");

#Customize the plot
ax.set(title="Heart Diesease and Cholestrol Level",
      xlabel="Age",
      ylabel="Cholestrol");

#Adding the legends
ax.legend(*scatter.legend_elements(),title="target");

#Adding a horizontal line
ax.axhline(over_50["chol"].mean(),
          linestyle='--');
```

```python
#Setting style to our plots
plt.style.use("seaborn-whitegrid")
#Subplot of age-thalach and age-chol together
fig,(ax0,ax1) = plt.subplots(nrows=2,
                            ncols=1,
                            figsize=(10,10),
                            sharex=True)
#sharex attribute help in all the subplots to share the same axes

#Providing data to the subplots
#Providing data to the ax0
scatter1=ax0.scatter(x=over_50["age"],
                   y=over_50["thalach"],
                   c=over_50["target"],
                   cmap="plasma")
#Customizing the ax0
ax0.set(title="Heart Diesease vs Thalach",
       xlabel="Heart Diesease",
       ylabel="Thalach")
#Setting the legend
ax0.legend(*scatter1.legend_elements(),title="target")

#Customizing mean Thalach line
ax0.axhline(over_50["thalach"].mean(),
           linestyle="--")

#Changing the x axis limits
ax0.set_xlim([50,80])

#Changing the y axis limits
ax0.set_ylim([80,200])

#Providing data to the subplots
#Providing data to the ax1
scatter2=ax1.scatter(x=over_50["age"],
                   y=over_50["chol"],
                   c=over_50["target"],
                    cmap="plasma")

#Customizing the ax1
ax1.set(title="Heart Diesease vs cholestrol",
       xlabel="Heart Diesease",
       ylabel="cholestrol")

#Setting the legend
ax1.legend(*scatter2.legend_elements(),title="target")

#Customizing mean cholestrol line
ax1.axhline(over_50["chol"].mean(),
           linestyle="--");

#Add title to the plot
fig.suptitle("Heart Disease Analysis",fontsize=18,fontweight=20);
```
